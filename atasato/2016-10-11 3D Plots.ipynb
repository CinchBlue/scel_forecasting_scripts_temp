```python
date = np.asarray(date)

date_uni = np.unique(date)

date1 = datetime.strptime(date_uni[0], "%m/%d/%Y")
date1 = date1.strftime('%Y%m%d')
date1 = np.asmatrix(date1)


for i in range(1,date_uni.size):
        datei = datetime.strptime(date_uni[i], "%m/%d/%Y")
        datei = datei.strftime('%Y%m%d')
        datei = np.asmatrix(datei)
        date1 = np.hstack((date1,datei))

date1 = np.asarray(date1)
date1 = np.unique(date1)
date1 = np.asmatrix(date1)
date1 = date1.T
print(date1)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
X = date1
Y = col_times
Z = col_irr

def get_min(time_str):
    h, m = time_str.split(':')
    return int(h) * 60 + int(m)

X = date1.A1
Y = col_times.apply(get_min).as_matrix()
Z = col_irr.apply(lambda x: x)

X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
