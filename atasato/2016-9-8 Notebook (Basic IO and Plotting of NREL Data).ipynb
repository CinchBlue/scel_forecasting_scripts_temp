```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from ipywidgets import interact, fixed
%matplotlib inline

print("Packages imported successfully.")
```

```python
# Data from: Southern Great Plains, Central Facility, Oklahoma
# Dates: 2013-6-25 to 2014-6-25
input_data = pd.read_csv("20030625.csv")

print(type(input_data))
print(input_data)
```

```python
data = input_data.ix[:, 2:4]

print(type(data))
print(data)
```

```python
# Plot of all data
plt.plot(data.ix[1:10000, ['Global PSP-A [W/m^2]', 'Global PSP-B [W/m^2]']])
```

```python
# Selected for t = 1 to 10000
plt.plot(data.ix[1:10000, ['Global PSP-A [W/m^2]', 'Global PSP-B [W/m^2]']])
```

```python
# Find total annual energy dispersed here in W * minute / m^2
sum_energy = sum(data.ix[:, 'Global PSP-A [W/m^2]'])
print('Sum of all recorded energy:       ', sum_energy)

# Divide all energy by 60 minutes per 1 hour. => W * h / m^2
annual_energy_per_hour = sum_energy / 60 / 1e6;
print("Total annual MWh/m^2:             ", annual_energy_per_hour)
```

```python
# Optimial energy: MWh/m^2
optimal_energy_per_hour = 1000 * 24 * 365 / 1e6
print("Optimal energy per hour: MWh/m^2: ", optimal_energy_per_hour)

# Capacity factor: experimental/optimal
capacity_factor = annual_energy_per_hour / optimal_energy_per_hour
print("Capacity factor:                  ", capacity_factor*100, "%")
```
