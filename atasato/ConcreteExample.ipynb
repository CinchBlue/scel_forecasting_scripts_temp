```python
#Load Libraries
from __future__ import division
import numpy as np 
import scipy as sp
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
from IPython.html.widgets import interact, fixed
%matplotlib inline

#Load Linear Regression Library
from sklearn.linear_model import LinearRegression

#Load data from UCI's Machine Learning Repository
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/concrete/compressive/Concrete_Data.xls'
concrete = pd.read_excel(url) 

#Make the X array
x = [concrete[concrete.columns[0]],concrete[concrete.columns[1]], concrete[concrete.columns[2]], concrete[concrete.columns[3]], concrete[concrete.columns[4]], concrete[concrete.columns[5]], concrete[concrete.columns[6]], concrete[concrete.columns[7]]]    
X = np.transpose(x)

#Make the Y array
Y = concrete[concrete.columns[8]]

#Initiatiate our linear regression object
model = LinearRegression(normalize = True )

#Load the X and Y array into the linear regression object
model.fit(X,Y)

#Print the coefficent and intercept
print "The coefficent are" , model.coef_

#Print the intercept
print "The intercept is" , model.intercept_

print type(model)
```

```python
from sklearn import cross_validation
from sklearn import neighbors
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, Y, test_size=0.3, random_state=0)
knn = neighbors.KNeighborsRegressor(n_neighbors=1)
knn.fit(X_train, y_train)
yhat_test=knn.predict(X_test)
plt.scatter(y_test,yhat_test)
plt.plot(y_test,y_test)
meanN1 = np.mean(y_test - yhat_test)
STDN1  = np.std(y_test - yhat_test)
print meanN1
print STDN1
```

```python
knn = neighbors.KNeighborsRegressor(n_neighbors=2)
knn.fit(X_train, y_train)
yhat_test=knn.predict(X_test)
plt.scatter(y_test,yhat_test)
plt.plot(y_test,y_test)
meanN5 = np.mean(y_test - yhat_test)
STDN5  = np.std(y_test - yhat_test)
print meanN5
print STDN5
```
